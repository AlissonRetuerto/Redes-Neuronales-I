```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Cargar el conjunto de datos Iris
iris = load_iris()

# Crear un DataFrame con AltoSepalo (x) y AnchoSepalo (y)
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Seleccionar los 20 primeros registros de AltoSepalo (x) y AnchoSepalo (y)
df_sepalos = df[['sepal length (cm)', 'sepal width (cm)']].head(20)

# Graficar los puntos
plt.figure(figsize=(8, 6))
plt.scatter(df_sepalos['sepal length (cm)'], df_sepalos['sepal width (cm)'], color='blue')
plt.title('Gráfico de Dispersión de AltoSepalo vs AnchoSepalo')
plt.xlabel('AltoSepalo (cm)')
plt.ylabel('AnchoSepalo (cm)')
plt.grid(True)
plt.show()
```

```python
# Crear un DataFrame con AltoPetalo (x) y AnchoPetalo (y)
df_petalos = df[['petal length (cm)', 'petal width (cm)']].head(20)

# Graficar los puntos
plt.figure(figsize=(8, 6))
plt.scatter(df_petalos['petal length (cm)'], df_petalos['petal width (cm)'], color='red')
plt.title('Gráfico de Dispersión de AltoPetalo vs AnchoPetalo')
plt.xlabel('AltoPetalo (cm)')
plt.ylabel('AnchoPetalo (cm)')
plt.grid(True)
plt.show()
```

```python
# Crear un histograma con AltoSepalo
plt.figure(figsize=(8, 6))
plt.hist(df['sepal length (cm)'], bins=10, color='green', edgecolor='black')
plt.title('Histograma de AltoSepalo')
plt.xlabel('AltoSepalo (cm)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos desde el archivo CSV
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vSkcRJ_3MlxwU3ZsfbTgJkZtzb0C_rKvvHCd7GCyqZDeyRfLApJWRiQc1TRO6J5KEonHzOdYaDHFDBL/pub?gid=1939740177&single=true&output=csv')

# 1. Mostrar columnas, tipos de datos, primeras y últimas 10 filas
print("\n▶️ Columnas:")
print(df.columns)
print("\n▶️ Tipos de datos:")
print(df.dtypes)
print("\n▶️ Primeras 10 filas:")
print(df.head(10))
print("\n▶️ Últimas 10 filas:")
print(df.tail(10))
```

```python
pasajero_148 = df.loc[147]

print("\n▶️ Datos del pasajero con identificador 148:\n")
print(pasajero_148)
```

```python
filas_pares = df.iloc[::2]

print("\n▶️ Filas pares del DataFrame:\n")
print(filas_pares)
```

```python
primera_clase = df[df['Pclass'] == 1]
nombres_ordenados = primera_clase['Name'].sort_values()
print("\n▶️ Nombres de personas en primera clase ordenados alfabéticamente:\n")
for nombre in nombres_ordenados:
    print(nombre)
```

```python
print("\n▶️ Porcentaje de supervivientes y fallecidos:")
survival_counts = df['Survived'].value_counts(normalize=True) * 100
print(survival_counts)

plt.figure(figsize=(6, 6))
plt.pie(survival_counts, labels=['Murieron', 'Sobrevivieron'], autopct='%1.1f%%', colors=['red', 'green'])
plt.title('Supervivencia en el Titanic')
plt.show()
```

```python
print("\n▶️ Porcentaje de supervivientes por clase:")
survived_by_class = df[df['Survived'] == 1]['Pclass'].value_counts(normalize=True) * 100
print(survived_by_class)

# Gráfico de pastel
plt.figure(figsize=(6, 6))
plt.pie(survived_by_class, labels=[f'Clase {i}' for i in survived_by_class.index], autopct='%1.1f%%')
plt.title('Supervivencia por Clase')
plt.show()
```

```python
df_sin_edad_nan = df.dropna(subset=['Age'])
print("\n▶️ DataFrame sin pasajeros con edad desconocida (NaN):\n")
print(df_sin_edad_nan)
```

```python
mujeres = df[df['Sex'] == 'female']

mujeres = mujeres.dropna(subset=['Age'])

edad_media_por_clase = mujeres.groupby('Pclass')['Age'].mean()

print("\n▶️ Edad media de las mujeres por clase:\n")
print(edad_media_por_clase)
```
