4.	Crea una función con Python que calcule la factorial de un número

```python
def factorial(n):
    if n < 0:
        return "El factorial no está definido para números negativos"
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado

numero = int(input("Ingrese un número entero: "))
print(f"El factorial de {numero} es {factorial(numero)}")
```

5.	Calcular detalle de liquidación de empleado. Según los datos ingresados empleado, sueldo básico, años de servicio y número de hijos. Calcular:
a.	Liquidación: 65% de sueldo por años de servicio
b.	CTS: 30% de liquidación
c.	Asignación Familiar: x hijos el 5% de sueldo por años de servicio
d.	d. Neto: Liquidación + CTS + Asignación Familiar


```python
def calcular_liquidacion():

    nombre = input("Ingrese el nombre del empleado: ")
    sueldo_basico = float(input("Ingrese el sueldo básico: "))
    anios_servicio = int(input("Ingrese los años de servicio: "))
    num_hijos = int(input("Ingrese el número de hijos: "))

    liquidacion = 0.65 * sueldo_basico * anios_servicio
    cts = 0.30 * liquidacion
    asignacion_familiar = 0.05 * sueldo_basico * anios_servicio * num_hijos
    sueldo_neto = liquidacion + cts + asignacion_familiar

    print("\n--- Detalle de Liquidación ---")
    print(f"Empleado: {nombre}")
    print(f"Sueldo Básico: S/ {sueldo_basico:.2f}")
    print(f"Años de Servicio: {anios_servicio}")
    print(f"Número de Hijos: {num_hijos}")
    print(f"Liquidación (65% x año): S/ {liquidacion:.2f}")
    print(f"CTS (30% de Liquidación): S/ {cts:.2f}")
    print(f"Asignación Familiar (5% x hijo x año): S/ {asignacion_familiar:.2f}")
    print(f"Sueldo Neto: S/ {sueldo_neto:.2f}")

calcular_liquidacion()
```

6.	El cálculo del pago mensual de un empleado de una empresa se efectúa de la siguiente manera: el sueldo básico se calcula en base al número total de horas trabajadas basado en una trifa horaria; al sueldo, se le aplica una bonificación del 20% obteniéndose el sueldo bruto; al sueldo bruto, se le aplica un descuento del 10% obteniéndose el sueldo neto. Escriba un programa que calcule e imprima el sueldo básico, el sueldo bruto y el sueldo neto de un trabajado

```python
def calcular_pago_mensual():

    nombre = input("Ingrese el nombre del empleado: ")
    horas_trabajadas = float(input("Ingrese el número total de horas trabajadas: "))
    tarifa_horaria = float(input("Ingrese la tarifa por hora: "))

    sueldo_basico = horas_trabajadas * tarifa_horaria
    bonificacion = sueldo_basico * 0.20
    sueldo_bruto = sueldo_basico + bonificacion
    descuento = sueldo_bruto * 0.10
    sueldo_neto = sueldo_bruto - descuento

    print("\n--- Información de Pago Mensual ---")
    print(f"Empleado: {nombre}")
    print(f"Sueldo Básico: S/ {sueldo_basico:.2f}")
    print(f"Bonificación (20%): S/ {bonificacion:.2f}")
    print(f"Sueldo Bruto: S/ {sueldo_bruto:.2f}")
    print(f"Descuento (10%): S/ {descuento:.2f}")
    print(f"Sueldo Neto: S/ {sueldo_neto:.2f}")

calcular_pago_mensual()
```

7.	Débora, Raquel y Sofia aportan cantidades de dinero para formar un capital. Diseñe un programa que determine el capital formado y el porcentaje de dicho capital que aporta cada uno

```python
def calcular_aportes():

    debora = float(input("Ingrese el aporte de Débora: S/ "))
    raquel = float(input("Ingrese el aporte de Raquel: S/ "))
    sofia = float(input("Ingrese el aporte de Sofía: S/ "))


    capital_total = debora + raquel + sofia


    porc_debora = (debora / capital_total) * 100
    porc_raquel = (raquel / capital_total) * 100
    porc_sofia = (sofia / capital_total) * 100


    print("\n--- Detalles del Capital y Porcentajes ---")
    print(f"Capital total formado: S/ {capital_total:.2f}")
    print(f"Débora aporta: S/ {debora:.2f} ({porc_debora:.2f}%)")
    print(f"Raquel aporta: S/ {raquel:.2f} ({porc_raquel:.2f}%)")
    print(f"Sofía aporta: S/ {sofia:.2f} ({porc_sofia:.2f}%)")

calcular_aportes()
```

8.	Una institución social tiene un centro de salud, un comedor infantil, una escuela infantil, y un asilo de ancianos. La institución recibe anualmente una donación que lo reporte de la siguiente forma: 25% de la donación para la implementación del centro de salud, 35% de la donacion para el comedor infantil, 25% de la donacion para le escuela infantil y el resto para el asilo de ancianos. Diseñe un algoritmo para efectuar el reparto de la donacion

```python
def repartir_donacion():

    donacion = float(input("Ingrese el monto total de la donación anual: S/ "))

    centro_salud = donacion * 0.25
    comedor_infantil = donacion * 0.35
    escuela_infantil = donacion * 0.25
    asilo_ancianos = donacion - (centro_salud + comedor_infantil + escuela_infantil)

    print("\n--- Reparto de la Donación ---")
    print(f"Centro de Salud (25%): S/ {centro_salud:.2f}")
    print(f"Comedor Infantil (35%): S/ {comedor_infantil:.2f}")
    print(f"Escuela Infantil (25%): S/ {escuela_infantil:.2f}")
    print(f"Asilo de Ancianos (15%): S/ {asilo_ancianos:.2f}")

repartir_donacion()
```

10.	Crea una clase en POO que contenga un método que devuelva el mayor de 2 números

```python
class Comparador:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def mayor(self):

        if self.num1 > self.num2:
            return self.num1
        else:
            return self.num2

numero1 = float(input("Ingresar el primer número: "))
numero2 = float(input("Ingresar el segundo número: "))


comparador = Comparador(numero1, numero2)

print(f"El mayor de los dos números es: {comparador.mayor()}")
```
